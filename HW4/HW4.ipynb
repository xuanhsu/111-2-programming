```python
import requests
from bs4 import BeautifulSoup
response = requests.get(
    "https://www.notebookcheck.net/?&hide_date=1&typeArray[]=1&id=48")
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())  #輸出排版後的HTML內容
```

```python
import csv
titles = soup.find_all(class_='introa_large introa_review')
with open('links.csv', 'w', newline='', encoding='utf-8') as f:
            writer = csv.writer(f)
            for title in titles:
                href = title.get('href')
                print(href)
                writer.writerow([href])
```

```python
#試著先導出一篇文章內容
# 導入 模組(module) 
import requests 
# 導入 BeautifulSoup 模組(module)：解析HTML 語法工具
import bs4

# 文章連結
URL = "https://www.notebookcheck.net/Samsung-Galaxy-Book3-Pro-14-review-Light-office-notebook-with-20-hours-of-battery-life.717518.0.html"
# 設定Header與Cookie
my_headers = {'cookie': 'over18=1;'}
# 發送get 請求 到 ptt 八卦版
response = requests.get(URL, headers = my_headers)
#  把網頁程式碼(HTML) 丟入 bs4模組分析
soup = bs4.BeautifulSoup(response.text,"html.parser")
## 查找所有html 元素 抓出內容
main_container = soup.find(id='nbc_contentcolumns_1col')
# 把所有文字都抓出來
all_text = main_container.text
# 把整個內容切割透過 "-- " 切割成2個陣列
pre_text = all_text.split('--')[0]
    
# 把每段文字 根據 '\n' 切開
texts = pre_text.split('\n')
# 內容
text = '\n'.join(texts)
print(text)
```

```python
# 導入 模組(module) 
import requests 
# 導入 BeautifulSoup 模組(module)：解析HTML 語法工具
import bs4
import csv

# 開啟 CSV 檔案
with open('links.csv', newline='', encoding='utf-8-sig') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)

    # 以迴圈輸出每一列
    for row in rows:
        URL = row[0]
        
# 設定Header與Cookie
        my_headers = {'cookie': 'over18=1;'}
# 發送get 請求 到 ptt 八卦版
        response = requests.get(URL, headers = my_headers)
#  把網頁程式碼(HTML) 丟入 bs4模組分析
        soup = bs4.BeautifulSoup(response.text,"html.parser")
## 查找所有html 元素 抓出內容
        main_container = soup.find(id='nbc_contentcolumns_1col')
# 把所有文字都抓出來
        all_text = main_container.text
# 把整個內容切割透過 "-- " 切割成2個陣列
        pre_text = all_text.split('--')[0]
    
# 把每段文字 根據 '\n' 切開
        texts = pre_text.split('\n')
# 內容
        text = '\n'.join(texts)
        print(text)
        # 在這裡使用 url 進行操作
```
