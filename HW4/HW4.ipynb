```python
import requests
from bs4 import BeautifulSoup
response = requests.get(
    "https://www.notebookcheck.net/?&hide_date=1&typeArray[]=1&id=48")
soup = BeautifulSoup(response.text, "html.parser")
print(soup.prettify())  #輸出排版後的HTML內容
```

```python
import csv
titles = soup.find_all(class_='introa_large introa_review')
for title in titles: 
    href = title.get('href')
    print(href)
    with open('links.csv', 'w', newline='', encoding='utf-8') as f:
        writer = csv.writer(f)
        for title in titles:
            writer.writerow([title])
```
