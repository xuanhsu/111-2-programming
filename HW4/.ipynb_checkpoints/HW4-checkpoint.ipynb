```python
import requests
import bs4
import csv


def scrape_content(url):
    my_headers = {'cookie': 'over18=1;'}
    response = requests.get(url, headers=my_headers)
    soup = bs4.BeautifulSoup(response.text, "html.parser")
    main_container = soup.find(id='nbc_contentcolumns_1col')
    all_text = main_container.text
    pre_text = all_text.split('--')[0]
    texts = pre_text.split('\n')
    text = '\n'.join(texts)
    return text


def main():
    with open('links.csv', newline='', encoding='utf-8-sig') as csvfile:
        rows = csv.reader(csvfile)
        for row in rows:
            url = row[0]
            try:
                content = scrape_content(url)
                print(content)
                with open('content.txt', 'a') as txtfile:
                    txtfile.write(content)
            except requests.exceptions.RequestException as e:
                print(f"Error scraping content from {url}: {str(e)}")


if __name__ == "__main__":
    main()
```

```python
#輸出抓到的全部網址
# 導入 模組(module) 
import requests 
# 導入 BeautifulSoup 模組(module)：解析HTML 語法工具
import bs4
import csv

# 開啟 CSV 檔案
with open('links.csv', newline='', encoding='utf-8-sig') as csvfile:
    # 讀取 CSV 檔案內容
    rows = csv.reader(csvfile)

    # 以迴圈輸出每一列
    for row in rows:
        URL = row[0]
        
# 設定Header與Cookie
        my_headers = {'cookie': 'over18=1;'}
# 發送get 請求 到 notebookcheck
        response = requests.get(URL, headers = my_headers)
#  把網頁程式碼(HTML) 丟入 bs4模組分析
        soup = bs4.BeautifulSoup(response.text,"html.parser")
## 查找所有html 元素 抓出內容
        main_container = soup.find(id='content')
# 把所有文字都抓出來
        all_text = main_container.text
# 把整個內容切割透過 "-- " 切割成2個陣列
        pre_text = all_text.split('--')[0]
    
# 把每段文字 根據 '\n' 切開
        texts = pre_text.split('\n')
# 內容
        text = '\n'.join(texts)
    #存入txt檔
        print(text)
```

```python
import requests
import bs4
import csv

with open('links.csv', newline='', encoding='utf-8-sig') as csvfile:
    rows = csv.reader(csvfile)
    with open('content.txt', 'w', encoding='utf-8') as txtfile:
        for row in rows:
            URL = row[0]
            my_headers = {'cookie': 'over18=1;'}
            response = requests.get(URL, headers=my_headers)
            soup = bs4.BeautifulSoup(response.text, "html.parser")
            main_container = soup.find(id='content')
            all_text = main_container.text
            pre_text = all_text.split('--')[0]
            texts = pre_text.split('\n')
            text = '\n'.join(texts)
            txtfile.write(text + '\n')
            print(text)
```
