```python
#(3)有最高頻率的處理器
import pandas as pd

# 讀取 CSV 檔案，指定第一欄為索引
df = pd.read_csv('11th gen intel.csv', index_col=0)

# 指定要找最大值的欄位名稱
column_to_find_max = 'Clock rate_base'

# 找到最大值
max_value = df[column_to_find_max].max()

# 建立布林遮罩，選擇符合最大值的那列資料
mask = df[column_to_find_max] == max_value
max_rows = df[mask]

subset = max_rows['Model']

print(f"Fastest clock rate：{max_value}")
print("CPUs with fastest clock rate：")
print(subset)

```

```python
#(4)大於32個處理單元顯示晶片的處理器
import pandas as pd

# 讀取 CSV 檔案，指定第一欄為索引
df = pd.read_csv('11th gen intel.csv', index_col=0)

# 選擇符合條件的資料
df_filtered = df[df['IGPU_EUs'] > 32]

subset = df_filtered[['Model', 'IGPU_EUs']]


# 輸出符合條件的資料
print(subset )
```

```python
#(5)功號小於7瓦的處理器
import pandas as pd

# 讀取 CSV 檔案，指定第一欄為索引
df = pd.read_csv('11th gen intel.csv', index_col=0)

# 選擇符合條件的資料
df_filtered = df[df['TDP_Watts'] < 10]

subset = df_filtered[['Model', 'TDP_Watts']]


# 輸出符合條件的資料
print(subset )
```

```python
#(6)哪個時段發表最多處理器
import pandas as pd

# 讀取 CSV 檔案，指定第一欄為索引
df = pd.read_csv('11th gen intel.csv', index_col=0)

# 使用value_counts()方法，計算每個值出現的次數
counts = df['Launch'].value_counts()

# 找出出現次數最多的值，並輸出
most_common = counts.idxmax()
subset = df.loc[df['Launch'] == most_common, ['Model', 'Launch']]
print(subset)
```

```python
#(7)快取不小於12MB且核心數不小於6的處理器
import pandas as pd

# 讀取 CSV 檔案，指定第一欄為索引
df = pd.read_csv('11th gen intel.csv', index_col=0)

# 選擇符合條件的資料
df_filtered = df[(df['Cores'] >= 6) & (df['Smart cache_MB'] >= 12)]

subset = df_filtered[['Model', 'Cores','Smart cache_MB']]


# 輸出符合條件的資料
print(subset)
```

```python
#(9)哪些處理器是移動端專用的
import pandas as pd

# 讀取 CSV 檔案，指定第一欄為索引
df = pd.read_csv('11th gen intel.csv', index_col=0)

subset = df.loc[df['Socket'] != 'LGA 1200', ['Model']]
print(subset)
```

```python
#(10)那些處理器核心與執行續數量相同
import pandas as pd

# 讀取 CSV 檔案，指定第一欄為索引
df = pd.read_csv('11th gen intel.csv', index_col=0)


# 檢查 A 和 B 兩欄是否相等，如果相等則輸出 C 值
for idx, row in df.iterrows():
    if row['Cores'] == row['Threads']:
        print(row['Model'])
```
