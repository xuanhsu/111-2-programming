```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import json


headers = {'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/89.0.4389.90 Safari/537.36'}
url='https://www.sinya.com.tw/diy/api_prods'
```

```python
res=requests.get(url,headers=headers).text
jsondata=json.loads(res)
jsondata
```

```python
data = []
for i in jsondata:
    if i['price'] == "0" or i['stockText'] =="【補貨中】" :
        continue
    elif i['stockText'] == "":
        name = i['prod_name']
        price = i['price']
        ds = i['discount']
        data.append({'product': name, 'price': price, 'discount': ds})

# 將Python資料結構轉換成JSON格式並寫入檔案中
with open('sinya.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
import json
import pandas as pd

# 讀取JSON格式的檔案
with open('sinya.json', 'r', encoding='utf-8') as f:
    data = json.load(f)

# 將JSON格式資料轉換成DataFrame格式
df = pd.DataFrame(data)

# 將DataFrame格式的資料輸出成CSV檔案
df.to_csv('sinya.csv', encoding='utf-8-sig', index=False)

# 輸出DataFrame格式的資料
print(df)
```

```python
# 將DataFrame格式的資料輸出成Excel檔案
df.to_excel('sinya.xlsx', index=False)

# 輸出DataFrame格式的資料
print(df)
```
